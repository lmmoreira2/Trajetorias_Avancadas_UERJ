```python
import numpy as np
import rk4

pi = np.pi
```

```python
class Integrand():
    def __init__(self, m, E, M, U):
        self.m = m
        self.E = E
        self.M = M
        self.U = U

    def __call__(self, vec_r, t):
        r = vec_r[0]
        theta = vec_r[1]
        return np.array([-1*np.sqrt( 2 / self.m * ( self.E - self.U(r) ) - self.M**2 / self.m**2 / r**2 ), self.M / self.m / r**2])


class Potential():
    def __init__(self, m, M):
        self.m = m
        self.M = M

    def __call__(self, r):
        return -self.M*self.m / r




vec_r = np.array([20.0, 0.0])
t = 0.0
dt = 0.1
mass = 0.1
Mass = 1500.0
angular_momentum = 10.0
energy = -5

hist_r = [vec_r[0]]
hist_theta = [vec_r[1]]
hist_t = []



U = Potential(mass, Mass)
integrand = Integrand(mass, energy, angular_momentum, U)
```
